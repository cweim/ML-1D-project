```python
# Read data from the .txt file
file_path = "RU/train"  # Replace with the actual path
with open(file_path, "r", encoding="utf-8") as file:
    lines = file.readlines()

# Process lines and create the data structure
data = []
sentence = []
labels = []
for line in lines:
    line = line.strip()
    if line:  # Non-empty line
        parts = line.split()
        if len(parts) == 2:
            word, label = parts
            sentence.append(word)
            labels.append(label)
        else:
            print(f"Ignoring invalid line: {line}")
    else:  # Empty line indicates end of sentence
        if sentence:
            data.append((sentence, labels))
            sentence = []
            labels = []

# Add the last sentence if it's not empty
if sentence:
    data.append((sentence, labels))
```

```python
def word2features(sent, i):
    word = sent[i]['word']
    # Add more features as needed
    features = {
        'bias': 1.0,
        'word.lower()': word.lower(),
        'word[-3:]': word[-3:],
    }
    return features

def sent2features(sent):
    return [word2features(sent, i) for i in range(len(sent))]

def sent2labels(sent):
    return [token['label'] for token in sent]
```

```python
from sklearn_crfsuite import CRF



# Convert data into features and labels
X_train = [sent for sent, _ in data]
y_train = [labels for _, labels in data]

# Define and train the CRF model
crf = CRF(algorithm='lbfgs',
          c1=0.1,
          c2=0.1,
          max_iterations=100,
          all_possible_transitions=True)
crf.fit(X_train, y_train)
```

```python
# Read test data from the .txt file
test_file_path = "RU/dev.in"  # Replace with the actual path
with open(test_file_path, "r", encoding="utf-8") as file:
    test_lines = file.readlines()

# Process test lines and create test_data
test_data = []
test_sentence = []
for line in test_lines:
    line = line.strip()
    if line:  # Non-empty line
        parts = line.split()
        if len(parts) == 2:
            word, label = parts
            test_sentence.append(word)
        else:
            word = parts[0]
            test_sentence.append(word)
    else:  # Empty line indicates end of sentence
        if test_sentence:
            test_data.append((test_sentence, []))  # Empty labels for test data
            test_sentence = []

# Add the last sentence if it's not empty
if test_sentence:
    test_data.append((test_sentence, []))
 
# Convert test data into features
X_test = [sent for sent, _ in test_data]

# Predict using the trained model
y_pred = crf.predict(X_test)

# Print the predicted labels
for sentence, pred_labels in zip(X_test, y_pred):
    print(" ".join(sentence))
    print(" ".join(pred_labels))
    print()  # Empty line to separate sentences
```

```python
# Define a function to convert a sentence and its predicted labels to text format
def labels_to_text(sentence, pred_labels):
    lines = []
    for word, label in zip(sentence, pred_labels):
        lines.append(f"{word} {label}")
    return "\n".join(lines) + "\n\n"

# Convert the predicted labels back to text format
output_text = ""
for sentence, pred_labels in zip(X_test, y_pred):
    output_text += labels_to_text(sentence, pred_labels)

# Write the formatted text to a new file
output_file_path = "RU/dev.p4.out"  # Replace with the desired output file path
with open(output_file_path, "w", encoding="utf-8") as output_file:
    output_file.write(output_text)

print("Output written to:", output_file_path)
```
